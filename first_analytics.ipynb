```python

seasons = {'Winter': [], 'Spring': [], 'Summer': [], 'Autumn': []}


for index, row in traffic_day.iterrows():
    if row['Период'].month in (1, 2, 12):
        seasons['Winter'].append(row['Итого и средние'])
    elif row['Период'].month in (3, 4, 5):
        seasons['Spring'].append(row['Итого и средние'])
    elif row['Период'].month in (6, 7, 8):
        seasons['Summer'].append(row['Итого и средние'])
    else:
        seasons['Autumn'].append(row['Итого и средние'])
seasons['Winter'] = sum(seasons['Winter'])
seasons['Spring'] = sum(seasons['Spring'])
seasons['Summer'] = sum(seasons['Summer'])
seasons['Autumn'] = sum(seasons['Autumn'])
```

```python
seasons_2022 = {'Winter': [], 'Spring': [], 'Summer': [], 'Autumn': []}


for index, row in traffic_day.iterrows():
    if row['Период'].month in (1, 2, 12) and str(row['Период'].year) == '2022':
        seasons_2022['Winter'].append(row['Итого и средние'])
    elif row['Период'].month in (3, 4, 5) and str(row['Период'].year) == '2022':
        seasons_2022['Spring'].append(row['Итого и средние'])
    elif row['Период'].month in (6, 7, 8) and str(row['Период'].year) == '2022':
        seasons_2022['Summer'].append(row['Итого и средние'])
    elif row['Период'].month in (9,10, 11) and str(row['Период'].year) == '2022':
        seasons_2022['Autumn'].append(row['Итого и средние'])
seasons_2022['Winter'] = sum(seasons_2022['Winter'])
seasons_2022['Spring'] = sum(seasons_2022['Spring'])
seasons_2022['Summer'] = sum(seasons_2022['Summer'])
seasons_2022['Autumn'] = sum(seasons_2022['Autumn'])
```
