```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(df.value, model='additive', period =2*24) # 24 hours = 30 min * 2 * 24
```

```python
fig = result.plot()
```

```python
result = seasonal_decompose(df.value, model='additive', period =2*24*7) # a week
fig = result.plot()
```

```python
result = seasonal_decompose(df.value, model='additive', period =2*24*30) # a month
fig = result.plot()
```
